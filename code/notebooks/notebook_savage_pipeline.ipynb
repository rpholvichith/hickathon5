NaNs are set such as each row has at least one NaN.

```python
tiny_train_data_Xy['meteo_radiation_IR'].unique()
```

```python
train_data['meteo_radiation_IR'].unique()
```

```python
# Drop 'meteo_radiation_IR' because no value in the first 100_000 rows
tiny_train_data_Xy = tiny_train_data_Xy.drop(columns=['meteo_radiation_IR'])
```

```python
tiny_train_data_X = tiny_train_data_Xy.select_dtypes(include=['int64', 'float64'])
tiny_train_data_X.info()
```

```python
tiny_train_data_withoutNA = tiny_train_data_X.fillna(tiny_train_data_X.mean()).join(tiny_train_data['piezo_groundwater_level_category'])
test_data_Xy = test_data.select_dtypes(include=['int64', 'float64']).drop(columns=['meteo_radiation_IR'])
test_data_withoutNA = test_data_Xy.fillna(test_data_Xy.mean())
```

```python
import numpy as np

X_tiny = tiny_train_data_withoutNA.select_dtypes(include=['int64', 'float64']).to_numpy()
y_tiny = tiny_train_data_withoutNA['piezo_groundwater_level_category'].to_numpy()
X_test = test_data_withoutNA.select_dtypes(include=['int64', 'float64']).to_numpy()

print(X_tiny.shape)
print(y_tiny.shape)
print(X_test.shape)
```

```python
from sklearn.model_selection import train_test_split

X_tiny_train, X_tiny_test, y_tiny_train, y_tiny_test = train_test_split(X_tiny, y_tiny, test_size=0.2, random_state=42)
print(X_tiny_train.shape)
print(y_tiny_train.shape)
print(X_tiny_test.shape)
print(y_tiny_test.shape)
```

```python
print(y_tiny_train[:5])
print(y_tiny_test[:5])
```

## Model training and prediction

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression().fit(X_tiny_train, y_tiny_train)
```

```python
pred = logistic_regression.predict(X_tiny_test)

print(pred[:100])
```

```python
from sklearn.metrics import f1_score, confusion_matrix

print(f"Accuracy: {logistic_regression.score(X_tiny_test, y_tiny_test)}")
print(f"F1 score: {f1_score(y_tiny_test, pred, average=None)}")
print(confusion_matrix(y_tiny_test, pred, labels=tiny_train_data_withoutNA['piezo_groundwater_level_category'].unique()))
print(tiny_train_data_withoutNA['piezo_groundwater_level_category'].unique())
```

```python
confusion_matrix(y_tiny_test, pred, labels=tiny_train_data_withoutNA['piezo_groundwater_level_category'].unique())
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_tiny_test, pred, labels=tiny_train_data_withoutNA['piezo_groundwater_level_category'].unique()))
```
